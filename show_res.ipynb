## presets

```python
from pathlib import Path

experiment_name = "presets-cv"

workdir = Path.cwd() / "scoring_results_aggregated" / experiment_name
```

```python
from collections import defaultdict
import yaml
import re

pattern = r"([^[]+)\[preset='([^']+)'\]\.yaml"


dataset_wise_results = defaultdict(list)
for yaml_file in workdir.glob("*.yaml"):
    # Extract dataset name and preset name from the filename using regex
    
    file_name = yaml_file.name
    
    # Pattern to match: <dataset_name>[preset='<preset_name>'].yaml
    match = re.match(pattern, file_name)

    dataset_name = match.group(1)
    preset_name = match.group(2)

    results = {"preset": preset_name}
    with open(yaml_file, "r") as f:
        results = {**results, **yaml.safe_load(f)}

    dataset_wise_results[dataset_name].append(results)
```

```python
import pandas as pd 

dataset_name = "banking77"
df = pd.DataFrame(dataset_wise_results[dataset_name])
# Function to highlight the maximum value in each column
def highlight_max(s):
    '''
    Highlights the maximum value in a Series yellow.
    '''
    is_max = s == s.max()
    return ['background-color: yellow' if v else '' for v in is_max]

# Apply the highlighting function to the dataframe
# Only apply to numeric columns (excluding 'preset' column)
numeric_columns = df.select_dtypes(include='number').columns
styled_df = df.style.apply(highlight_max, subset=numeric_columns)

# Display the styled dataframe
styled_df
```
