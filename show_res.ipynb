## best scorers

эта визуализация показывает насколько хорошее качество выдают скореры

```python
from pathlib import Path

experiment_name = "holdout-v2"

workdir = Path.cwd() / "all_best_scorers_aggregated" / experiment_name
```

```python
datasets_seeds_pairs = list(path for path in workdir.iterdir())
```

```python
import pandas as pd


results = {name.stem: pd.read_csv(workdir / name) for name in datasets_seeds_pairs}
```

```python
# Get the dataframe
df = results["snips"]
df
# Create a styled dataframe with max values highlighted
styled_df = df.style.highlight_max(
    subset=[
        "mean_scoring_accuracy",
        "std_scoring_accuracy",
        "mean_scoring_f1",
        "std_scoring_f1",
        "mean_scoring_precision",
        "std_scoring_precision",
        "mean_scoring_recall",
        "std_scoring_recall",
        "mean_scoring_roc_auc",
        "std_scoring_roc_auc",
    ],
    color="yellow",
)

# Display the styled dataframe
styled_df
```
