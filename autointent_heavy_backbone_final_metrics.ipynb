```python
import json
from pathlib import Path

filepath = Path("wandb_final_scorers/scorers-ho-heavy-backbone/results.json")

with filepath.open("r") as f:
    data = json.load(f)

len(data)
```

```python
def extract_dataset_and_seed(key):
    """
    Extract dataset name and seed from a string like "banking77[seed='44']"
    
    Args:
        key (str): String in format "dataset_name[seed='seed_value']"
        
    Returns:
        tuple: (dataset_name, seed_value)
    """
    import re
    
    # Use regex to extract dataset name and seed value
    match = re.match(r"([^[]+)\[seed='(\d+)'\]", key)
    
    if match:
        dataset_name = match.group(1)
        seed = match.group(2)
        return dataset_name, seed
    else:
        return None, None
```

```python
from collections import defaultdict
grouped_res = defaultdict(list)

for group, res in data.items():
    dataset, seed = extract_dataset_and_seed(group)
    grouped_res[dataset].append(res[0]["config"]["value"])
```

```python
grouped_res
```

```python
rows = []
for dataset, metrics_dicts in grouped_res.items():
    metrics_list = defaultdict(list)
    for metrics_dict in metrics_dicts:
        for metric, value in metrics_dict.items():
            metrics_list[metric].append(value)
    mean_metrics = {metric: sum(values) / len(values) for metric, values in metrics_list.items()}
    rows.append({
        "dataset": dataset,
        **mean_metrics
    })
rows
```

```python
import pandas as pd

df = pd.DataFrame(rows)
df
```
